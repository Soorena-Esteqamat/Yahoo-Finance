```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36",
}
```

```python
url = "https://finance.yahoo.com/quote/ADA-USD/history/?period1=1588550400&period2=1732815268"
response = requests.get(url, headers = headers)
```

```python
response.ok
```

```python
soup = BeautifulSoup(response.text)
```

```python
rows = soup.find_all('tr', class_='yf-j5d1ld')[1:]
data = []
for row in rows:
    cells = row.find_all('td')
    row_data = [cell.text for cell in cells]
    data.append(row_data)

columns = ["Date", "Open", "High", "Low", "Close", "Adjusted Close", "Volume"]
df = pd.DataFrame(data, columns=columns)
df['Volume'] = df['Volume'].str.replace(',', '').astype(float)
columns_to_convert = ['Open', 'High', 'Low', 'Close', 'Adjusted Close']
df[columns_to_convert] = df[columns_to_convert].astype(float)
df
```

```python
df.to_csv('data.csv', index = False)
```
